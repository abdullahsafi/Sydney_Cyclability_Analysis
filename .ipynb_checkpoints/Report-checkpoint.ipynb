## 1.  Dataset Description

*What are your data sources and how did you obtain and pre-process the data?*
### Neighborhoods:
A CSV file consisting of attributes on unique areas within NSW. This datasource contains information on area characteristics such as size, as well as population and number of business. When inserting the data into postgres, we made sure to clean any empty values in land_area, population, number_dwellings with NaN.
### Statistical Areas:
Csv file which contains unique ids for different regions in Australia, as well as names and parent area ids.
### Business Stats:
A CSV file which comprises of business information attributed to each unique region/areas. This datasource includes information on the number of businesses, retail, accommodation and other services such as  arts and recreational, food, health, social  assistance, education. When inserting the data into postgres, we made sure to clean any empty values with 0, this is because an empty value in "education and training" for example means that there is 0 business involving education and training.
### Census Stats:
CSV file with financial information for each area, data includes household income and monthly rent. When inserting the data into postgres, we made sure to clean any empty values in median_income, and avg_monthly_rent with NaN.
### Bike Sharing Pods:
Another csv file with information on bike pods. Each entry is identified with a station id. Data includes the name of the pod, the number of bikes and scooters, geographic data such as latitude and longitude, and lastly a description of where the pods are located.
### Bike Parking Data:
A json file that contains information on the locations of bike parking spots along with the number of bikes it can hold. Location data is set by longitude and latitude values and the dataset contains an object_id value which we will use as a primary key. We retrived this json data by accessing a web API using the python request library (City of Sydney Council).
### SA2 Shapefile Geom Data:
Collected a shapefile from (Australian Bureau of Statistic) containing the boundaries of Sydney neighbourhoods. We proccessed this using a shapefile reader and inputed the bounderies into the Neighbourhoods table under the column geom.

## 2. Database Description

*Into which database schema did you integrate your data (preferable shown with a diagram)?
Which index(es) did you create, and why?*

The schema we chose is much aligned with the datasources, we represented Neighbourhoods, Business Statistics, Census Statistics, Statistical Areas in seperate tables in the database. Neighbourhoods is the primary table with primary key area_id. This key gets referenced by CensusStats, StatisticalAreas, BusinessStatas, zScores as a foreign key, in those tables, this foreign key constraint specifies that area_id can only contain values that are in referenced Neighbourhood table.

For the tables BikeSharingPods and BikeParking, we used the PostGIS ST_Contains function to conduct a spatial join of longitude and latitude point with the Neighbourhoods geometry polygon. This allows us to locate whether or not the longitude and latitude point of a BikeSharingPod and BikeParking is within a Neighbourhood.

We created an index on Neighbourhoods(area_id) "nhood_area", this is useful for all of the z-scores involving density calculations. Density calculations required population or land_area, this index supports the joins between the tables with Neighbourhoods. Likewise an Index was created on zScores(area_id) to increase the speed of our correlation analysis. Furthermore, a spatial index was created on Neighbourhoods(geom) using GiST in order to increase the speed of our spatial joins.

## 3. Cyclability Analysis

To calculate cyclability per neighbourhood, a formula was used that took all data sources with relevance to cyclability into consideration. The formula is as follows:

$cyclability = z(population density)+z(dwelling density)+z(service balance)+z(bikepod density)+z(BikeParking)$

Where $z$ is a z-score or "standard score", which is calculated for each of the attributes using this formula: 

$z(measure, x) = x  avg(measure) / stddev(measure)$
                    
                    
Population density, dwelling density and BikeParking are relative to each neighbourhoods land area. Bikepod density is the number of pods per suburb and is implemented through a spatial join. The service balance score was calculated by finding the ratio of selected businesses over the total number of businesses in an neighbourhood, the types of businesses that were selected are:
- retail trade
- accommodation and food services
- health care and social assistance

These types of businesses can be justified in assisting for a more accurate cyclability score because places that offer retail are more likely to contain bike parking spaces. Food and accommodation are attractions for cyclists as pit stops on trips and will increase an areas cyclability score. Lastly, health care is also a significant because nearby health care facilities can provide for injured cyclists, thus increasing cyclability score.

Despite this, it is important to consider the limitations of the cyclability score. Some of data is specific to certain regions in sydney, which may have altered results. This is because the cyclability score may be skewed higher towards these areas, as they contain more data to analyse and therefore will have higher scores. Moreover, in some of the data sources there are missing values, invalid values and null entries, which clearly can have an affect on the cyclability score. In addition to this, the cyclability score may not be as accurate, as there is not enough relevant and recent data to represent each suburb in sydney and across NSW. Nevertheless, these limitations should not have a significant influence over the outcome of the cyclability score and we can take our results into consideration as a true representation of NSWs cyclability.

```python
import matplotlib.pyplot as plt

#run usyd server postgres
import psycopg2
import psycopg2.extras

def pgconnect():
    # please replace <your_unikey> and <your_SID> with your own details
    YOUR_UNIKEY = 'bwin6658'#'<your_unikey>'
    YOUR_PW     = '480372747'#'<your_SID>'
    try: 
        conn = psycopg2.connect(host='soit-db-pro-1.ucc.usyd.edu.au',
                                database='y19s1d2x01_'+YOUR_UNIKEY,
                                user='y19s1d2x01_'+YOUR_UNIKEY,
                                password=YOUR_PW) 
        print('connected')
    except Exception as e:
        print("unable to connect to the database")
        print(e)
        return None
    return conn

conn = pgconnect()

def pgquery( conn, sqlcmd, args=None, msg=False, returntype='tuple'):
    """ utility function to execute some SQL query statement
        it can take optional arguments (as a dictionary) to fill in for placeholders in the SQL
        will return the complete query result as return value - or in case of error: None
        error and transaction handling built-in (by using the 'with' clauses)"""
    retval = None
    with conn:
        cursortype = None if returntype != 'dict' else psycopg2.extras.RealDictCursor
        with conn.cursor(cursor_factory=cursortype) as cur:
            try:
                if args is None:
                    cur.execute(sqlcmd)
                else:
                    cur.execute(sqlcmd, args)
                if (cur.description != None ):
                    retval = cur.fetchall() # we use fetchall() as we expect only _small_ query results
                if msg != False:
                    print("success: " + msg)
            except psycopg2.DatabaseError as e:
                if e.pgcode != None and msg:
                    print("db read error: "+msg)
                    print(e)
            except Exception as e:
                print(e)
    return retval
```

```python
import numpy as np
query = """SELECT area_name, cyclability
           FROM Neighbourhoods n Join zScores s ON (n.area_id = s.area_id)
           Order By cyclability DESC
           Limit 10"""
all_scores = pgquery(conn, query, returntype='dict')

area_names = []
cyclability_scores = []

for area in all_scores:
    area_names.append(area['area_name'])
    cyclability_scores.append(area['cyclability'])

index = np.arange(len(area_names))
plt.bar(index, cyclability_scores)
plt.xticks(index, area_names, fontsize=10, rotation=80)
plt.title("10 Best Cycling Areas")
plt.xlabel("Area Names")
plt.ylabel("Cyclability Scores")
plt.show()
    
```

As presented in the figure above, the top ten best cycling areas are grouped by their respective cyclability scores. Areas such as Woolloomooloo, Darlinghurst, Surry Hill and Redfern stand out the most. This is mostly likely due to the fact that they are suitable places to cycle relative to the rest of NSWs suburbs, since areas such as these have many bikepod stations, an abundance of businesses that accommodate cyclists and also provide the most bike parking. In addition to this, dwelling and population density are suitable for smaller areas such as these, and therefore most people would rather cycle or walk to their destination, which facilitates an area to be more cyclist friendly. Moreover, according to the Roads and Maritimes Services [cycleway finder](https://www.rms.nsw.gov.au/maps/cycleway_finder) <sup>[1]</sup> , these areas contain the most bike paths. Therefore, it is conspicuous that we can conclude that these results are seemingly valid.

## 4. Correlation Analysis

*How well does your score correlate to the median household income in the given suburbs?
Is there any correlation with the average weekly rent in the neighbourhoods?*

<div id='id-section4.1'/>

### Cyclability and Average Monthly Rent

```python
query = """SELECT cyclability, avg_monthly_rent
           FROM CensusStats Join zScores using (area_id)
           Where avg_monthly_rent != double precision 'NaN'"""
result = pgquery(conn, query, returntype='dict')

cyclability_scores = []
average_rent = []

for area in result:
    cyclability_scores.append(float(area['cyclability']))
    average_rent.append(float(area['avg_monthly_rent']))


plt.scatter(average_rent,cyclability_scores)
plt.title("Correlation between Cyclability and Average Monthly Rent")
plt.xlabel("Average Monthly Rent")
plt.ylabel("Cyclability Scores")
z = np.polyfit(average_rent, cyclability_scores, 1)
p = np.poly1d(z)
plt.plot(average_rent,p(average_rent),"r--")
plt.text(750,10,"Correlation = 0.279")
plt.show()
```


The graph above displays the cyclability scores plotted against the average monthly rent. At a glance, the data does seem to represent somewhat of a weak trend. As the average monthly rent increases, the cyclability score also increases. This trend is most apparent in the 1500 - 3000 range, since a majority of high cyclability scores come from areas that pay a higher average monthly rent.  


<div id='id-section4.2'/>

### Cyclability and Median Annual Household Income

```python
query = """SELECT cyclability, median_annual_household_income
           FROM CensusStats Join zScores using (area_id)
           Where median_annual_household_income != double precision 'NaN'"""
result = pgquery(conn, query, returntype='dict')

cyclability_scores = []
median_annual_household_income = []

for area in result:
    cyclability_scores.append(float(area['cyclability']))
    median_annual_household_income.append(float(area['median_annual_household_income']))


plt.scatter(median_annual_household_income,cyclability_scores)
plt.title("Correlation between Cyclability and Median Annual Household Income")
plt.xlabel("Median Annual Household Income")
plt.ylabel("Cyclability Scores")
z = np.polyfit(median_annual_household_income, cyclability_scores, 1)
p = np.poly1d(z)
plt.plot(median_annual_household_income,p(median_annual_household_income),"r--")
plt.text(30000,10,"Correlation = 0.213")
plt.show()
```

This graph does not present any evident and clear trend. Most of the cyclability scores are grouped towards the 40,000 to 60,000 income range. However, it is apparent that areas with an annual household income of 60,000+ experience higher cyclability scores relative to the rest. There are some outliers around the 50,000 range and the best range for cyclability seems to be around the 60,000 mark. 

<div id='id-section4.2'/>

### Statistical Results

The cyclability score of each neighbourhood was tested against the areas average weekly rent and median household income. Overall, the correlation coefficients calculated are 0.279 and 0.213 respectively. 

The results of our correlation calculations represent a somewhat weak positive linear relationship, and therefore the analysis concluded that our cyclability score was not completely correlated to median household income and weekly rent. 

The cyclability score has a weak correlation with median household income and average weekly rent because when constructing the cyclability score, factors that define the wealth of a suburb were not heavily taken into consideration. For potential future projects other significant factors such as bike lanes, road size and traffic may help to identify expensive/wealthier areas and therefore if there is a true correlation between these areas and cyclability score. Despite this, the cyclability score can still be applied, even if the the analysis concludes that it is not correlated to household income and weekly rent.
