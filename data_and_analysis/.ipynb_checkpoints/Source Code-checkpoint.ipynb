```python
#create function to make queries to postgres database

def pgquery( conn, sqlcmd, args=None, msg=False, returntype='tuple'):
    """ utility function to execute some SQL query statement
        it can take optional arguments (as a dictionary) to fill in for placeholders in the SQL
        will return the complete query result as return value - or in case of error: None
        error and transaction handling built-in (by using the 'with' clauses)"""
    retval = None
    with conn:
        cursortype = None if returntype != 'dict' else psycopg2.extras.RealDictCursor
        with conn.cursor(cursor_factory=cursortype) as cur:
            try:
                if args is None:
                    cur.execute(sqlcmd)
                else:
                    cur.execute(sqlcmd, args)
                if (cur.description != None ):
                    retval = cur.fetchall() # we use fetchall() as we expect only _small_ query results
                if msg != False:
                    print("success: " + msg)
            except psycopg2.DatabaseError as e:
                if e.pgcode != None and msg:
                    print("db read error: "+msg)
                    print(e)
            except Exception as e:
                print(e)
    return retval
```

Setting up cleaning data function

```python
import numpy as np
DEFAULT_VALUE = np.nan
DEFAULT_STRING = ""

def clean(data, column_key, convert_function, default_value):
    special_values= {} # no special values yet
    for row in data:
        old_value = row[column_key]
        new_value = default_value
        try:
            if old_value in special_values.keys():
                new_value = special_values[old_value]
            else:
                new_value = convert_function(old_value)
        except (ValueError, TypeError):
            print('Replacing {} with {} in column {}'.format(row[column_key], new_value, column_key))
        row[column_key] = new_value
```

## Insert Neighbourhoods CSV Files Into Database

Also adding SA2 Area Shapefile to Geom Column

```python
import csv
import re
import shapefile
import time
sf = shapefile.Reader("1270055001_sa2_2016_aust_shape/SA2_2016_AUST.shp", encoding="iso-8859-1")
shapes = sf.shapes()
records= sf.records()
neighbourhood_data = list(csv.DictReader(open('Neighbourhoods.csv')))

for row in neighbourhood_data:
    area_name = row['area_name']
    
    for j in range(len(shapes)):
        if records[j][2] == area_name:
            row['geom']="POLYGON(("
            i=0
            for x, y in shapes[j].points:
                row['geom']+="%s %s," % (x,y)
               # check for start of a new polygon part
                i += 1
                if i in shapes[j].parts:
                    row['geom']= re.sub(",$", "),(", row['geom'])
            # properly end the polygon string
            row['geom'] = re.sub(",$", "))", row['geom'])
```

```python
clean(neighbourhood_data, 'area_id', str.strip, DEFAULT_VALUE)
clean(neighbourhood_data, 'area_name', str.strip, DEFAULT_VALUE)

# the following converts the two measurment  columns to float  values - or NaN
clean(neighbourhood_data, 'land_area', float, DEFAULT_VALUE)
clean(neighbourhood_data, 'population', float, DEFAULT_VALUE)
clean(neighbourhood_data, 'number_of_dwellings', float, DEFAULT_VALUE)
clean(neighbourhood_data, 'number_of_businesses', float, DEFAULT_VALUE)
```

```python
#create new table in postgres
neighbourhood_schema = '''CREATE TABLE Neighbourhoods (
                     area_id CHAR(9) PRIMARY KEY, 
                     area_name VARCHAR(80), 
                     land_area NUMERIC, 
                     population NUMERIC, 
                     number_of_dwellings NUMERIC,
                     number_of_businesses NUMERIC,
                     geom GEOMETRY(Polygon,4326))''' 

#pgquery(conn, "CREATE EXTENSION postgis;", msg="added postgis")
pgquery(conn, "DROP TABLE Neighbourhoods", msg="cleared old table")
pgquery(conn, neighbourhood_schema, msg="created Neighbourhoods table")
```

### Calculating bike pod density z-score for each neighbourhood

```python
#get the counts of bike pods per area
bike_pod_query = """SELECT N.area_id, Count(*)
                      FROM Neighbourhoods N JOIN BikeSharingPods B ON ST_Contains(N.geom, ST_SetSRID(ST_MakePoint(B.longitude, B.latitude), 4326))
                      Group By N.area_id
                   """
bike_pod_result = pgquery(conn, bike_pod_query, returntype='dict')
pprint(bike_pod_result[1])
```

```python
#create new table in postgres for bike pod density
BikePodDensity_schema = '''CREATE TABLE BikePodDensity (
                     area_id CHAR(9), 
                     count NUMERIC, 
                     density NUMERIC )''' 

pgquery(conn, "DROP TABLE BikePodDensity", msg="cleared old BikePodDensity table")
pgquery(conn, BikePodDensity_schema, msg="created BikePodDensity table")
```

```python
for neighbour in neighbourhoods:
    values = {}
    values['area_id'] = neighbour['area_id']
    land_area = neighbour['land_area']
    values['bike_pods'] = 0
    for area in bike_pod_result:
        if values['area_id'] == area['area_id']:
            values['bike_pods'] = area['count']
            break
    if land_area and values['bike_pods'] > 0:
        values['density'] = values['bike_pods']/land_area
        insrt_stmt = "INSERT INTO BikePodDensity VALUES (%(area_id)s, %(bike_pods)s, %(density)s)"
        pgquery(conn, insrt_stmt, args = values, msg="inserted")
    else:
        values['density'] = 0
        insrt_stmt = "INSERT INTO BikePodDensity VALUES (%(area_id)s, %(bike_pods)s, %(density)s)"
        pgquery(conn, insrt_stmt, args = values, msg="inserted")
```

```python
query = "SELECT area_id, density FROM BikePodDensity"
count = pgquery(conn, query, returntype='dict')
#pprint(count)
```

```python
query = """SELECT AVG(density) as "avg", stddev(density) as "std"
           FROM BikePodDensity;"""
bike_density = pgquery(conn, query, returntype='dict')

pprint(bike_density)

average_bike_density = bike_density[0]['avg']
std_bike_density = bike_density[0]['std']
```

```python
query = """SELECT n.area_id, (b.density - """ + str(average_bike_density) +""")/ """ + str(std_bike_density) +""" as "z_score"
           FROM Neighbourhoods n Join BikePodDensity b Using (area_id);"""
bike_density_scores = pgquery(conn, query, returntype='dict')

#pprint(bike_density_scores)
```

```python
for area in bike_density_scores:
    z_score_dict[area['area_id']]['bike_density'] = round(area['z_score'], 5)
    
#print(z_score_dict)
```

## Insert z-scores into postgres

```python
#create new table in postgres
Scores_schema = '''CREATE TABLE zScores (
                     area_id CHAR(9), 
                     bike_pods_z NUMERIC, 
                     pop_density_z NUMERIC, 
                     dwelling_density_z NUMERIC,
                     service_balance_z NUMERIC,
                     bike_parking_z NUMERIC,
                     cyclability NUMERIC)''' 

#pgquery(conn, "CREATE EXTENSION postgis;", msg="added postgis")
pgquery(conn, "DROP TABLE zScores", msg="cleared old Scores table")
pgquery(conn, Scores_schema, msg="created Scores table")
```

```python
import math

for area in z_score_dict:
    row = {}
    row['cyclability'] = 0
    row['area_id'] = area
    row['bike_pods_z'] = z_score_dict[area]['bike_density']
    row['cyclability'] += float(row['bike_pods_z'])
    row['bike_parking_z'] = z_score_dict[area]['bike_parking_density']
    row['cyclability'] += float(row['bike_parking_z'])
    try:
        row['pop_density_z'] = z_score_dict[area]['pop_density']
        row['cyclability'] += float(row['pop_density_z'])
    except:
        row['pop_density_z'] = math.nan
        
    try:
        row['dwelling_density_z'] = z_score_dict[area]['dwelling_density']
        row['cyclability'] += float(row['dwelling_density_z'])
    except:
        row['dwelling_density_z'] = math.nan
        
    try:
        row['service_balance_z'] = z_score_dict[area]['service_balance']
        row['cyclability'] += float(row['service_balance_z'])
    except:
        row['service_balance_z'] = math.nan
    
    query = "INSERT INTO zScores VALUES(%(area_id)s,%(bike_pods_z)s,%(pop_density_z)s,%(dwelling_density_z)s, %(service_balance_z)s, %(bike_parking_z)s, %(cyclability)s )"
    pgquery(conn, query, args = row, msg="inserted "+ str(area))
```

```python
query = """SELECT area_name, cyclability, bike_parking_z
           FROM Neighbourhoods n Join zScores s ON (n.area_id = s.area_id)
           Order By cyclability DESC"""
all_scores = pgquery(conn, query, returntype='dict')
pprint(all_scores)
```

## Find correlation of neighbourhood cyclability score and average monthly rent per 

```python
query = """SELECT corr(cyclability , avg_monthly_rent)
           FROM CensusStats Join zScores using (area_id)
           Where avg_monthly_rent != double precision 'NaN'"""
result = pgquery(conn, query, returntype='dict')

pprint(result)
```

## Find correlation of neighbourhood cyclability score and median household income

```python
query = """SELECT corr(cyclability , median_annual_household_income)
           FROM CensusStats Join zScores using (area_id)
           Where median_annual_household_income != double precision 'NaN'"""
result = pgquery(conn, query, returntype='dict')

pprint(result)
```
